```python
from datasets import load_dataset
```

```python
dataset = load_dataset('code_search_net', 'python')
```

```python
dataset
```

```python
from tqdm import tqdm

repos = set()

for x in dataset['train']['repository_name']:
  repos.add(x)
for x in dataset['test']['repository_name']:
  repos.add(x)
for x in dataset['validation']['repository_name']:
  repos.add(x)

print(len(repos))
```

```python
for x in list(repos)[:10]:
  print(x)
```

```python
def save(repos, out_file):
  with open(out_file, 'w') as f:
    f.write('\n'.join(repos))
```

```python
tests = [f"https://github.com/{repo}/tree/master/docs" for repo in repos]
tests[:4]
```

```python
import requests

has_docs = []
no_docs = []

for url in tqdm(tests):
  x = requests.get(url)
  if x.status_code == 200:
    has_docs.append(url)
    #print(url, ' has docs')
  else:
    no_docs.append(url)
    #print(url, ' no docs')

print()
print(len(has_docs))
print(len(no_docs))
```

```python
url_with_docs = [x[:-17] for x in has_docs]
save(url_with_docs, 'cosesearchnet_python_repos_with_docs_folder.txt')
url_with_docs[:5]
```

```python
all_urls = [f"https://github.com/{repo}" for repo in repos]
save(all_urls, "codesearchnet_python_repos.txt")
all_urls[:5]
```
